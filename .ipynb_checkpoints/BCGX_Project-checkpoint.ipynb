# BCG X Virtual Experience – Task 2: Exploratory Data Analysis (EDA)

In this task, we aim to understand PowerCo's customer churn problem by exploring the client and price datasets. Our goal is to identify patterns, correlations, and potential drivers of churn using Python-based exploratory data analysis (EDA).

---

## Step 1: Load and Preview the Data

We start by importing the required libraries and loading the datasets `client_data.csv` and `price_data.csv`. The `client_df` contains customer-level features, while `price_df` contains pricing history.




```python
import pandas as pd

# Load datasets
client_df = pd.read_csv("client_data.csv")
price_df = pd.read_csv("price_data.csv")

# Inspect structure
print("Client Data:")
display(client_df.info())
display(client_df.head())

print("\nPrice Data:")
display(price_df.info())
display(price_df.head())
```

```python
# Descriptive statistics for client data
client_df.describe()
```

```python
# Descriptive statistics for price data
price_df.describe()
```

## Step 2: Descriptive Statistics

We calculated descriptive statistics to explore the range and distribution of numeric features.

**Key observations:**
- Electricity consumption (`cons_12m`) and last month's consumption (`cons_last_month`) show high variance, with many customers reporting very low usage.
- Forecast-related variables (`forecast_cons_12m`, `forecast_cons_year`) also show wide ranges, indicating diverse customer expectations.
- Margin-related features (`margin_gross_pow_ele`, `net_margin`) vary significantly, which may be predictive of churn.
- The `price_data.csv` file shows that many price values are zero, especially in peak and mid-peak periods, suggesting non-applicability or missing values.

These insights indicate high variability in customer behavior and suggest potential features for modeling.

---

```python
# Churn and other categorical features
print("Churn Distribution:")
display(client_df['churn'].value_counts())

print("\nGas Subscription (has_gas):")
display(client_df['has_gas'].value_counts())

print("\nSales Channels:")
display(client_df['channel_sales'].value_counts())
```

## Step 3: Categorical Variable Distributions

We examined the distribution of key categorical features:

- **Churn**: 1,419 customers (≈9.7%) churned, while 13,187 stayed. This imbalance will be important when building models.
- **has_gas**: The majority of customers (≈82%) do not have a gas subscription, while ≈18% do.
- **channel_sales**: Sales channels are heavily skewed, with the top two categories making up the majority of entries. Several records are marked as "MISSING".

This step highlighted the need to handle class imbalance and possibly encode or consolidate sparse categorical variables.

---


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Churn vs Electricity Consumption (cons_12m)
plt.figure(figsize=(8, 5))
sns.histplot(data=client_df, x='cons_12m', hue='churn', bins=50, kde=True, element='step')
plt.title('Electricity Consumption (12m) by Churn')
plt.xlabel('cons_12m')
plt.ylabel('Count')
plt.show()

# Churn vs Forecasted Consumption (forecast_cons_12m)
plt.figure(figsize=(8, 5))
sns.histplot(data=client_df, x='forecast_cons_12m', hue='churn', bins=50, kde=True, element='step')
plt.title('Forecasted Consumption (12m) by Churn')
plt.xlabel('forecast_cons_12m')
plt.ylabel('Count')
plt.show()

# Churn vs Gross Margin on Power (margin_gross_pow_ele)
plt.figure(figsize=(8, 5))
sns.histplot(data=client_df, x='margin_gross_pow_ele', hue='churn', bins=50, kde=True, element='step')
plt.title('Gross Power Margin by Churn')
plt.xlabel('margin_gross_pow_ele')
plt.ylabel('Count')
plt.show()

# Churn vs Has Gas
plt.figure(figsize=(6, 4))
sns.barplot(x='has_gas', y='churn', data=client_df, estimator=lambda x: sum(x)/len(x))
plt.title('Churn Rate by Gas Subscription')
plt.xlabel('has_gas')
plt.ylabel('Churn Rate')
plt.show()
```

## Step 4: Visual Exploration of Churn Drivers

We visualized several variables to examine how they relate to customer churn:

- **Electricity Consumption (`cons_12m`)**: Churned customers tend to have lower consumption compared to those who stayed.
- **Forecasted Consumption (`forecast_cons_12m`)**: A similar pattern is observed—customers with lower expected consumption are more likely to churn.
- **Gross Margin on Power (`margin_gross_pow_ele`)**: Lower margins are associated with higher churn, suggesting that less profitable customers are harder to retain.
- **Gas Subscription (`has_gas`)**: Customers with a gas subscription churn less often, indicating that bundled services improve retention.

These visualizations reveal consistent trends: low-usage, low-margin customers, and those without bundled services are more prone to churn.

---

## Summary of Findings

- The churn rate is relatively low but significant enough to warrant predictive modeling.
- Churners generally have lower electricity usage, forecasted demand, and margins.
- Bundled service customers (with gas) show better retention.
- The sales channel variable is sparse and may need consolidation or encoding.
- Many price features contain zero values; these may require imputation or careful handling.
